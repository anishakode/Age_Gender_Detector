### Arthimetic Operation On Images

```python
img2 = cv2.imread("Resources/tint1.jpg", 0)
```

```python
# Addition
result = cv2.resize(img1, (200,200)) + img2
plt.imshow(result)
```

```python
# Subtarction
result = cv2.resize(img1, (200,200)) - img2
plt.imshow(result)
```

```python
# Multiplication
result = cv2.resize(img1, (200,200)) * img2
plt.imshow(result)
```

```python
# Division
result = cv2.resize(img1, (200,200)) / img2
plt.imshow(result)
```

### Bluring

```python
img1 = cv2.imread("Resources/brain.png")
img = cv2.cvtColor(img1, cv2.COLOR_BGR2HSV)
figure_size=7
result=cv2.blur(img, (figure_size, figure_size))
plt.figure(figsize=(11,6))
plt.subplot(121), plt.imshow(cv2.cvtColor(img, cv2.COLOR_HSV2RGB)), plt.title("Original")
plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(cv2.cvtColor(result, cv2.COLOR_HSV2RGB)), plt.title("Blur Image")
plt.xticks([]), plt.yticks([])
plt.show
```

### Rezize

```python
(height,width) = img1.shape[0:2]
zoom = cv2.resize(img1, None, fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

```python
zoom = cv2.resize(img1, (int(height*1.7), width*2), interpolation=cv2.INTER_CUBIC)
plt.imshow(zoom)
```

### Rotate

```python
m = cv2.getRotationMatrix2D((width/2,height/2),-30,1)
rotate = cv2.warpAffine(img1, m, (width, height))
plt.imshow(rotate)
```

### Translate

```python
a = np.float32([[1,0,50],[0,1,100]])
translate = cv2.warpAffine(img1, a, (width,height))
plt.imshow(translate)
```
